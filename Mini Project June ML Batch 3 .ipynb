```python
## Average number of words in movie titles between the year 2000-2005

##YEAR 2000

y=df[df['release_year'] ==2000]
  
l=[]
for i in y.original_title:
    e=i.replace(" ",'')
    l.append(len(e))

    
print(sum(l)/len(l))
```
